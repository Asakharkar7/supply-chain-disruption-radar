```python
df = spark.table("`workspace`.`default`.`container_ship_clean_delta`")
df.printSchema()
df.show(5)
```

```python
from pyspark.sql.functions import (
    hour, dayofweek, month, col, unix_timestamp, when
)

df = df \
    .withColumn("hour_of_day", hour("UPDATED_TS")) \
    .withColumn("day_of_week", dayofweek("UPDATED_TS")) \
    .withColumn("month", month("UPDATED_TS")) \
    .withColumn("sched_vs_actual_eta_hours",
                (unix_timestamp("ATA_TS") - unix_timestamp("ETA_SCHEDULE_TS")) / 3600)
```

```python
from pyspark.sql.window import Window
import pyspark.sql.functions as F

w = Window.partitionBy("SHIP").orderBy("UPDATED_TS")

df = df \
    .withColumn("prev_sog", F.lag("SOG").over(w)) \
    .withColumn("prev_hdg", F.lag("HDG").over(w)) \
    .withColumn("speed_change", col("SOG") - col("prev_sog")) \
    .withColumn("heading_change", F.abs(col("HDG") - col("prev_hdg")))
```

```python
import math
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType

def haversine(lat1, lon1, lat2, lon2):
    if None in (lat1, lon1, lat2, lon2):
        return None
        
    R = 6371  # km
    dlat = math.radians(lat2 - lat1)
    dlon = math.radians(lon2 - lon1)
    lat1 = math.radians(lat1)
    lat2 = math.radians(lat2)
    
    a = (math.sin(dlat/2)**2 +
         math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2)
    c = 2 * math.asin(math.sqrt(a))
    return R * c

haversine_udf = udf(haversine, DoubleType())

df = df \
    .withColumn("prev_lat", F.lag("LAT").over(w)) \
    .withColumn("prev_long", F.lag("LONG").over(w)) \
    .withColumn("distance_km", haversine_udf(col("prev_lat"), col("prev_long"), col("LAT"), col("LONG")))
```

```python
from pyspark.sql.window import Window
import pyspark.sql.functions as F

df = df.withColumn("SHIP_idx",     F.dense_rank().over(Window.orderBy("SHIP")))
df = df.withColumn("DEPPORT_idx",  F.dense_rank().over(Window.orderBy("DEPPORT")))
df = df.withColumn("ARRPORT_idx",  F.dense_rank().over(Window.orderBy("ARRPORT")))
df = df.withColumn("IMO_idx",      F.dense_rank().over(Window.orderBy("IMO")))

```

```python
feature_cols = [
    "hour_of_day", "day_of_week", "month",
    "SOG", "speed_change", "heading_change", "distance_km",
    "SHIP_idx", "DEPPORT_idx", "ARRPORT_idx", "IMO_idx",
]

X = pdf[feature_cols]
y_reg = pdf["DELAY_HOURS"]
y_cls = pdf["DELAY_CATEGORY"]
```

 Model 1: ETA Delay Regression (Random Forest)

```python
import numpy as np
import pandas as pd

X = X.fillna(X.median(numeric_only=True))
y_reg = y_reg.fillna(y_reg.median())

```

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(
    n_estimators=80,
    max_depth=10,
    n_jobs=-1,
    random_state=42
)

rf_reg.fit(X, y_reg)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

pred_reg = rf_reg.predict(X)

mae = mean_absolute_error(y_reg, pred_reg)
r2 = r2_score(y_reg, pred_reg)

print("MAE (hours):", mae)
print("R Score:", r2)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier

# encode ON_TIME / MODERATE / SEVERE
le = LabelEncoder()
y_cls_enc = le.fit_transform(pdf["DELAY_CATEGORY"])

rf_cls = RandomForestClassifier(
    n_estimators=150,
    max_depth=12,
    n_jobs=-1,
    random_state=42
)

rf_cls.fit(X, y_cls_enc)
```

```python
from sklearn.metrics import classification_report

pred_cls = rf_cls.predict(X)
print(classification_report(y_cls_enc, pred_cls))
```

```python
from pyod.models.iforest import IForest

iso = IForest(contamination=0.02, random_state=42)
iso.fit(X)

pdf["anomaly_score"] = iso.decision_function(X)
pdf["is_anomaly"] = iso.predict(X)
```
